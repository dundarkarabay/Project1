```python
import pandas as pd
```

```python
# Reading full Austin crime data from a csv file (dowloaded via API previously) 
df = pd.read_csv("../../Desktop/FullAustinCrimeReport.csv")
df.tail()
```

```python
df.columns
```

```python
# This first data frame will contain occurance date and time info for crimes. 
# We'll use the year, month,day and time info in our analysis.

occ_date_df = df.drop(['Unnamed: 0', ':@computed_region_8spj_utxs',
                       ':@computed_region_a3it_2a2z', ':@computed_region_q9nd_rr82',
                       ':@computed_region_qwte_z96m', 'address', 'category_description',
                       'census_tract', 'clearance_date', 'clearance_status',
                       'council_district', 'crime_type', 'district', 'family_violence',
                       'incident_report_number', 'latitude', 'location', 'location_type',
                       'longitude', 'occ_date_time', 'pra', 'rep_date',
                       'rep_date_time', 'rep_time', 'sector', 'ucr_category', 'ucr_code',
                       'x_coordinate', 'y_coordinate', 'zip_code'], axis=1)
```

```python
# This step runs ~ for 3 minutes.
# We're extracting the year, month and day to seperate columns as we'll use them in our analysis.

year_list = []
month_list = []
day_list = []

for index,row in occ_date_df.iterrows():
    occurance_date = row["occ_date"].split(sep="T")
    date = occurance_date[0].split(sep="-")
    year_list.append(date[0])
    month_list.append(date[1])
    day_list.append(date[2])
dictionary = {"occ_year": year_list,
              "occ_month": month_list,
              "occ_day": day_list}
addition_df = pd.DataFrame(dictionary)
```

```python
occ_date_df = occ_date_df.reset_index()
addition_df = addition_df.reset_index()
occ_date_df = pd.merge(occ_date_df,addition_df,how="left",on="index")
occ_date_df = occ_date_df.drop(["occ_date",'index'],axis=1)
```

```python
occ_date_df.to_csv("../../Desktop/FullAustinCrime_occ_year_month_day_Report.csv", header=True, index=False)
```

```python
# Deleting the columns that are either not necessary or contained in other columns (Ex: (1) rep_time and rep_date 
# are already available in rep_data_time which is kept. Ex: (2) Info given in location column is already available 
# in latitude and longitude columns)

short_df = df.drop(['census_tract', 'latitude', 'longitude', 'ucr_code',
                    'Unnamed: 0',':@computed_region_8spj_utxs', ':@computed_region_a3it_2a2z',
                    ':@computed_region_q9nd_rr82', ':@computed_region_qwte_z96m', 'x_coordinate', 
                    'y_coordinate','category_description','incident_report_number','occ_date',
                    'rep_date','rep_time','address', 'location','ucr_category','sector','pra','district'],axis=1)
short_df.tail()
```

```python
# Dropping the data that contains at least one "NaN".

clean_short_df = short_df.dropna()
clean_short_df.tail()
```

```python
clean_short_df.count()
```

```python
# Resetting the index and dropping the former index column 
clean_short_df = clean_short_df.reset_index()
clean_short_df = clean_short_df.drop(["index"],axis=1)
clean_short_df.tail()
```

```python
# This step runs ~ for 3 minutes.
# We're extracting the year, month and day to seperate columns as we'll use them in our analysis.

year_list = []
month_list = []
day_list = []

for index,row in clean_short_df.iterrows():
    occ_date_time = row["occ_date_time"].split(sep="T")
    occ_date = occ_date_time[0].split(sep="-")
    year_list.append(occ_date[0])
    month_list.append(occ_date[1])
    day_list.append(occ_date[2])
dictionary = {"occ_year": year_list,
              "occ_month": month_list,
              "occ_day": day_list}
addition_df = pd.DataFrame(dictionary)
```

```python
clean_short_df = clean_short_df.reset_index()
addition_df = addition_df.reset_index()
clean_short_df = pd.merge(clean_short_df,addition_df,how="left",on="index")
```

```python
clean_short_df = clean_short_df.drop(['index', 'occ_date_time'],axis=1)
clean_short_df.columns
```
